```python
df = pd.read_csv("clean_data.csv")
df.shape
```

```python
target_col = 'Approved_Flag'

X=df.drop(columns=[target_col])
y=df[target_col]
```

```python
education_map = {
    'SSC' : 1,
    '12TH' : 2,
    'GRADUATE' : 3,
    'UNDER GRADUATE' : 4,
    'PROFESSIONAL' : 3,
    'POST-GRADUATE' : 4,
    'OTHERS' : 1
}

X['EDUCATION'] = X['EDUCATION'].map(education_map).astype(int)
```

```python
#one-hot encoding
nominal_cols = [
    'MARITALSTATUS',
    'GENDER',
    'last_prod_enq2',
    'first_prod_enq2'
]

X_encoded = pd.get_dummies(
    X,
    columns=nominal_cols,
    drop_first = True
)
```

```python
from numpy._core import numeric
from sklearn.preprocessing import StandardScaler
numerical_cols = X_encoded.select_dtypes(exclude='object').columns.tolist()

scaler = StandardScaler()
X_encoded[numerical_cols] = scaler.fit_transform(X_encoded[numerical_cols])
```

#Credit risk intuition checks:
- Higher delinquency → higher default risk
- Higher income → lower default risk
- Longer employment duration → lower risk
- Recent payment delays → higher risk
- No feature violated basic credit monotonicity assumptions.


```python
print("total features before encoding:",X.shape[1])
print("total features after encoding:",X_encoded.shape[1])
```

```python
X_encoded['Approved_Flag'] = y
X_encoded.to_csv("model_ready_data.csv",index=False)
```
